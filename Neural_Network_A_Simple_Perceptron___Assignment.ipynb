**Question 1: What is Deep Learning? Briefly describe how it evolved and how it differs from traditional machine learning.**

**Answer:**

What is Deep Learning?
Deep learning is a subfield of machine learning that uses algorithms inspired by the human brain's neural networks to learn from vast amounts of data. Instead of requiring human-engineered features, deep learning models automatically learn a hierarchy of features directly from raw data. For example, in an image recognition task, a deep learning model might learn to recognize edges and shapes in the initial layers and then use those learned features to identify more complex objects like faces or cars in later layers.

**Evolution:**

* 1943: McCulloch–Pitts neuron model.
* 1958: Rosenblatt’s Perceptron.
* 1980s: Backpropagation algorithm.
* 2012: AlexNet breakthrough in computer vision.
* Today: Deep CNNs, RNNs, Transformers for text, speech, images.

**Difference from traditional ML:**

1. **Deep Learning:**

* **Architecture:** Employs deep neural networks with multiple hidden layers.
* **Data:** Requires large volumes of raw data and excels with unstructured data like images, audio, and text.
* **Feature Engineering:** Automatically learns complex features and patterns directly from the data, reducing the need for manual feature engineering.
* **Complexity:** Handles complex, non-linear relationships and tasks.
* **Interpretability:** Models are often complex and difficult to interpret due to their intricate structure.
* **Applications:** Ideal for image recognition, natural language processing, and autonomous systems.

2. **Traditional Machine Learning:**

* **Architecture:** Uses various algorithms such as decision trees, random forests, and support vector machines.
* **Data:** Performs well with structured, tabular data and requires less data than deep learning.
* **Feature Engineering:** Requires human expertise to extract and engineer relevant features from the data before training the model.
* **Complexity:** Generally deals with simpler, more linear correlations.
* **Interpretability:** Models are often simpler and easier to understand and interpret.
* **Applications:** Effective for tasks like regression, classification, and clustering with structured data.

**Question 2: Explain the basic architecture and functioning of a Perceptron. What are its limitations?**

**Answer:**

A perceptron's architecture includes input nodes, weights, a bias, a summation function, and a hard-limiting activation function to produce a binary output. It functions by taking inputs, multiplying them by their respective weights, adding a bias, and then passing the sum through the activation function to classify data. Its primary limitations are the inability to solve non-linearly separable problems, like the XOR problem, and its restriction to binary outputs.

**Perceptron Architecture:**

The basic architecture of a perceptron consists of the following components:     
* **Input Layer:** Receives input values (features) from the dataset.
* **Weights:** Each input is associated with a weight, representing its importance to the output.
* **Bias:** A fixed value added to the weighted sum of inputs, which helps to shift the decision boundary.
* **Summation Function:** Calculates the weighted sum of the inputs and adds the bias: \(\Sigma =(x_{1}w_{1}+x_{2}w_{2}+...+x_{n}w_{n})+b\).
* **Activation Function:** A hard-limiting function (e.g., step function) that produces a binary output (0 or 1) based on whether the weighted sum exceeds a certain threshold.

**Perceptron Functioning**

* **Input Presentation:** Input features (\(x_{1},x_{2},...,x_{n}\)) are fed into the perceptron.
* **Weighted Sum:** Each input is multiplied by its corresponding weight (\(w_{1},w_{2},...,w_{n}\)), and these products are summed, with the bias term added: \(sum=\sum _{i=1}^{n}(x_{i}w_{i})+b\).
* **Activation:** The result of the summation is then passed through the activation function. If the sum is above a threshold, the output is 1; otherwise, it is 0.
* **Output:** The perceptron generates a binary (0 or 1) output, representing a class.

**Limitations of the Perceptron:**

* **Linear Separability:** A perceptron can only classify linearly separable data, meaning data points that can be separated by a single straight line or hyperplane.
* **Cannot Solve XOR Problem:** It cannot solve non-linearly separable problems, such as the classic XOR (exclusive OR) function, because its decision boundary is not a straight line.
* **Binary Output:** The hard-limiting activation function restricts the output to binary values (0 or 1), limiting its ability to make probabilistic predictions.
* **Single-Layered Architecture:** Because it has only a single layer, the perceptron lacks the capability to model complex patterns that require more intricate decision boundaries, which are found in multi-layer networks.

**Question 4: What is the difference between Loss function and Cost function in neural networks? Provide examples.**

**Answer:**

In neural networks, a loss function quantifies the error for a single training example, while a cost function is the average of these loss values over the entire dataset or a batch, providing an overall performance measure for the model. For example, the Mean Squared Error (MSE) is often used as a loss function in regression, and the cost function would be the average MSE across all data points.

**Loss Function:**

* **Definition:** A function that measures the difference or error between the predicted output and the actual target value for one specific training data point.
* **Purpose:** To provide immediate feedback on how well the model is performing on an individual example.

**Cost Function:**

* **Definition:** An aggregate function that represents the overall performance of the model across the entire training dataset. It is essentially the average or sum of the loss functions computed for every single data point.
* **Purpose:** To provide a single, comprehensive metric that the model aims to minimize during training to improve overall performance.

**Key Difference:**

The core difference is scope: the loss function provides a fine-grained measure for one instance, while the cost function gives a broad, holistic evaluation across all instances. The model's training process aims to minimize the cost function, which is calculated using the individual losses from the loss function.

**Question 5: What is the role of optimizers in neural networks? Compare Gradient Descent, Adam, and RMSprop.**

**Answer:**

Optimizers in neural networks are algorithms that adjust model weights to minimize the loss function, enabling the model to learn and converge to optimal solutions. Gradient Descent is a fundamental method that takes steps in the direction of the steepest descent of the loss function. Adam combines the benefits of RMSprop and momentum, adaptively adjusting the learning rate and converging faster. RMSprop adapts the learning rate per parameter by dividing it by an exponentially decaying average of squared gradients, helping with non-stationary objectives and avoiding overshooting.

**Role of Optimizers**

* **Weight Adjustment:** Optimizers update the weights and biases of a neural network to find the best combination that minimizes the loss function.
* **Convergence:** They guide the training process, helping the model converge to a minimum loss, which indicates that the model has learned effectively.
* **Performance:** Optimizers influence convergence speed, can prevent divergence or oscillations, and can affect how well the model generalizes to unseen data.

**Comparison of Optimizers**

1. **Gradient Descent (and its variants, e.g., Stochastic Gradient Descent - SGD)**

* **Mechanism:** Iteratively updates model parameters in the direction that reduces the loss. SGD is a variant that updates parameters after each training example, offering a balance between speed and noise.
* **Pros:** Simple to understand and implement.
* **Cons:** Can be slow, especially for complex loss landscapes, and may get stuck in local minima or saddle points.

2. **RMSprop (Root Mean Square Propagation)**

* **Mechanism:** Adjusts the learning rate for each parameter by dividing it by an exponentially decaying average of past squared gradients.
* **Pros:**
   * Adapts the learning rate, helping to prevent large steps in steep regions and allowing larger steps in flatter regions.
   * Effective for non-stationary objectives and recurrent neural networks (RNNs).
* **Cons:** Does not incorporate momentum, which can lead to slower convergence in some cases.

3. **Adam (Adaptive Moment Estimation)**

* **Mechanism:** Combines the advantages of both momentum and RMSprop. It keeps exponentially decaying averages of both the gradients and their uncentered variance.
* **Pros:**
   * Often converges faster than other optimizers like RMSprop and SGD.
   * Handles both dense and sparse features well.
   * Performs bias correction for more stable early learning.
* **Cons:** While highly effective, it is generally a more complex algorithm than simpler gradient descent methods.

**Key Differences Summarized**

* **Learning Rate Adaptation:** Adam and RMSprop are adaptive, adjusting learning rates per parameter, while standard Gradient Descent uses a fixed learning rate.
* **Momentum:** Adam incorporates momentum to accelerate gradients in consistent directions, while RMSprop does not, though it uses an adaptive learning rate to control overshooting.
* **Complexity & Performance:** Adam is more computationally intensive but often yields superior results, making it a popular default choice for many applications.

**Question 6: Write a Python program to implement a single-layer perceptron from scratch using NumPy to solve the logical AND gate.**

(Include your Python code and output in the code box below.)

**Answer:**

```python
import numpy as np

# AND gate inputs & outputs
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])

# Add bias term
Xb = np.c_[np.ones((X.shape[0],1)), X]

weights = np.random.rand(3)
lr = 0.1

for epoch in range(20):
    for i in range(len(y)):
        z = np.dot(weights, Xb[i])
        pred = 1 if z > 0 else 0
        error = y[i] - pred
        weights += lr * error * Xb[i]

print("Final weights:", weights)
for i in range(len(y)):
    print(X[i], "->", 1 if np.dot(weights,Xb[i])>0 else 0)
```

**Question 7: Implement and visualize Sigmoid, ReLU, and Tanh activation functions using Matplotlib.**

(Include your Python code and output in the code box below.)

Answer:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)

sigmoid = 1 / (1 + np.exp(-x))
relu = np.maximum(0, x)
tanh = np.tanh(x)

plt.figure(figsize=(8,6))
plt.plot(x, sigmoid, label='Sigmoid', linewidth=2)
plt.plot(x, relu,   label='ReLU',    linewidth=2)
plt.plot(x, tanh,   label='Tanh',    linewidth=2)

plt.title("Activation Functions", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```

Output:
The graph shows the three activation functions—Sigmoid, ReLU, and Tanh—plotted across the input range [−10,10].

**Question 8: Use Keras to build and train a simple multilayer neural network on the MNIST digits dataset. Print the training accuracy.**

(Include your Python code and output in the code box below.)

**Answer:**

```python
import tensorflow as tf
from tensorflow import keras

# Load and normalize MNIST dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

# Build the model
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
history = model.fit(x_train, y_train, epochs=5, batch_size=32, verbose=1)

# Print the final training accuracy
print("Final Training Accuracy:", history.history['accuracy'][-1])
```

**Question 9: Visualize the loss and accuracy curves for a neural network model trained on the Fashion MNIST dataset. Interpret the training behavior.**

(Include your Python code and output in the code box below.)

**Answer:**

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras

# 1️⃣ Load & normalize Fashion MNIST
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()
x_train, x_test = x_train/255.0, x_test/255.0

# 2️⃣ Build a simple neural network
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

# 3️⃣ Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 4️⃣ Train with validation split to track validation curves
history = model.fit(x_train, y_train,
                    epochs=10,
                    batch_size=32,
                    validation_split=0.2,
                    verbose=1)

# 5️⃣ Plot training & validation loss/accuracy curves
plt.figure(figsize=(10,5))

# Loss
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Loss Curve')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# Accuracy
plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title('Accuracy Curve')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

**Interpretation of Training Behavior**

* Smoothly decreasing training & validation loss: Model is learning effectively.
* Training and validation accuracy both rising and converging: Good generalization.
* If training accuracy ≫ validation accuracy or validation loss rises while training loss keeps falling: Overfitting is occurring.
* If both losses plateau early: Model may be underfitting; consider more epochs or a deeper network.

**Question 10: You are working on a project for a bank that wants to automatically detect fraudulent transactions. The dataset is large, imbalanced, and contains structured features like transaction amount, merchant ID, and customer location. The goal is to classify each transaction as fraudulent or legitimate.**

**Explain your real-time data science workflow:**

* How would you design a deep learning model (perceptron or multilayer NN)?
* Which activation function and loss function would you use, and why?
* How would you train and evaluate the model, considering class imbalance?
* Which optimizer would be suitable, and how would you prevent overfitting?

(Include your Python code and output in the code box below.)

**Answer:**

**Real-Time Data Science Workflow**

1. **Data Understanding & Preprocessing**

* **Data cleaning:** Handle missing values, outliers, encode categorical variables (e.g. merchant ID, location) using one-hot or embeddings.
* **Feature scaling:** Normalize numerical features like transaction amount.
* **Train/validation/test split:** Use a stratified split so the rare fraud class is represented.

2. **Model Design**

* **Architecture:**

   * **Input layer:** size = number of features after preprocessing.
   * **Hidden layers:** 2–3 Dense layers (e.g. 64 → 32 neurons) with ReLU activation.
   * **Output layer:** 1 neuron with Sigmoid activation to output fraud probability.

* **Why ReLU + Sigmoid?**

   * ReLU avoids vanishing gradients in hidden layers.
   * Sigmoid outputs probability [0,1] suitable for binary classification.

3. **Loss Function**

* **Binary cross-entropy:** Ideal for two-class problems with probabilistic outputs.
* To handle class imbalance, pass class_weight to model.fit or oversample the minority class (SMOTE) as an alternative.

4. **Optimizer**

* **Adam:** Adaptive learning rate, usually converges faster and needs minimal tuning.

5. **Training & Evaluation**

* **Metrics:** Use AUC/ROC or F1-score besides accuracy (accuracy can be misleading on imbalanced data).
* **Callbacks:** EarlyStopping to avoid unnecessary epochs once validation loss stops improving.

6. **Preventing Overfitting**

* Dropout layers (e.g., 0.3–0.5).
* L2 (ridge) regularization on Dense layers.
* Early stopping and proper validation split.

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Example: assume you already have a NumPy feature matrix X and labels y
# X: shape (n_samples, n_features), y: 0 = legitimate, 1 = fraud

# Generate some dummy data for demonstration
# Ensure each class has more than one sample for stratification
n_samples = 1000
n_features = 10
X = np.random.rand(n_samples, n_features)
y = np.random.randint(0, 2, n_samples)

# To ensure the 'fraud' class (1) has at least 2 samples for stratification
if np.sum(y) < 2:
    y[0:2] = 1

# 1. Split & scale
X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val   = scaler.transform(X_val)

# 2. Build model
model = keras.Sequential([
    keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(1, activation='sigmoid')
])

# 3. Compile with binary cross-entropy and Adam
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy', keras.metrics.AUC(name='auc')])

# 4. Handle class imbalance
# Example: if fraud class is only 5% of data
# Dynamically calculate class weights based on the generated data
total_samples = len(y_train)
class_counts = np.bincount(y_train)
class_weight = {i: total_samples / (len(np.unique(y_train)) * count) for i, count in enumerate(class_counts)}

# 5. Train with EarlyStopping
early_stop = keras.callbacks.EarlyStopping(
    monitor='val_loss', patience=5, restore_best_weights=True
)

history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=256,
    validation_data=(X_val, y_val),
    class_weight=class_weight,
    callbacks=[early_stop],
    verbose=1
)

# 6. Evaluate
results = model.evaluate(X_val, y_val, verbose=0)
print(f"Validation Accuracy: {results[1]:.4f}")
print(f"Validation AUC: {results[2]:.4f}")
```

**Interpretation of Typical Output**

* **Validation Accuracy:** May be high (e.g., 0.95) but can be misleading if the dataset is highly imbalanced.
* **Validation AUC:** Better measure; values closer to 1.0 indicate good discrimination between fraudulent and legitimate transactions.
* Loss and accuracy curves should flatten without large divergence between train and validation sets—showing good generalization.
